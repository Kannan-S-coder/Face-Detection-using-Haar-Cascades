# FACE DETECTION

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img)

    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        cv2.imshow("Face detection",face_img)
    return face_img
```

```python
result1=detect_face(with_glass)
plt.imshow(result1,cmap='gray')
plt.show()
```

```python
result1=detect_face(with_out_glass)
plt.imshow(result1,cmap='gray')
plt.show()
```

```python
result1=detect_face(grp_photo)
plt.imshow(result1,cmap='gray')
plt.show()
```

# EYES DETECTION

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    eye_img = img.copy()
    eye_rect = eye_cascade.detectMultiScale(eye_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in eye_rect:
        cv2.rectangle(eye_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
        
    return eye_img
```

```python
result1=detect_eyes(with_glass)
plt.imshow(result1,cmap='gray')
plt.show()
```

```python
result1=detect_eyes(with_out_glass)
plt.imshow(result1,cmap='gray')
plt.show()
```

```python
result1=detect_eyes(grp_photo)
plt.imshow(result1,cmap='gray')
plt.show()
```

# Conjuction with Video

```python
cap = cv2.VideoCapture(0)

# Set up matplotlib
plt.ion()
fig, ax = plt.subplots()

ret, frame = cap.read()
frame = detect_face(frame) 

im=ax.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
plt.title('Video Face Detection')

# While True
while True:
    ret, frame = cap.read(0) 
    frame = detect_face(frame)
    
    # Update matplotlib image
    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)) # The image shows im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    
    plt.pause(0.10)

cap.release()
plt.close()

```
